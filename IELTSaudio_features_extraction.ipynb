```python
class Word:
    ''' A class representing a word from the JSON format for vosk speech recognition API '''

    def __init__(self, dict):
        '''
        Parameters:
          dict (dict) dictionary from JSON, containing:
            conf (float): degree of confidence, from 0 to 1
            end (float): end time of the pronouncing the word, in seconds
            start (float): start time of the pronouncing the word, in seconds
            word (str): recognized word
        '''

        self.conf = dict["conf"]
        self.end = dict["end"]
        self.start = dict["start"]
        self.word = dict["word"]

    def to_string(self):
        ''' Returns a string describing this instance '''
        return "{:20} from {:.2f} sec to {:.2f} sec, confidence is {:.2f}%".format(
            self.word, self.start, self.end, self.conf*100)
```

```python
import os
import sys
import wave
import json

from vosk import Model, KaldiRecognizer, SetLogLevel

SetLogLevel(0)
model_path = "/content/drive/MyDrive/Colab Notebooks/ML /Speaking IELTS project/models/vosk-model-en-us-0.22"
model = Model(model_path)
```

```python
#reading file
audio_filename="AU03818_2.wav"
wf = wave.open(audio_filename, "rb")

#recognize
rec = KaldiRecognizer(model, wf.getframerate())
rec.SetWords(True)
results = []

# recognize speech using vosk model
while True:
    data = wf.readframes(4000)
    if len(data) == 0:
        break
    if rec.AcceptWaveform(data):
        part_result = json.loads(rec.Result())
        results.append(part_result)

part_result = json.loads(rec.FinalResult())
results.append(part_result)
```

```python
import pandas as pd
data = pd.DataFrame(results)
data=pd.concat([data.drop(['result'], axis=1), data['result'].apply(pd.Series)], axis=1)
data=data.drop(['text'], axis=1)
data=data.stack().reset_index()
data=data.drop(['level_0','level_1'], axis=1)
data=pd.concat([data.drop([0], axis=1), data[0].apply(pd.Series)], axis=1)
data=data.sort_values('start')
data['hesitation']=data['start']-data['end'].shift(1)
data
```
